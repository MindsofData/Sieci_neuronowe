### <a name='a3'></a>Podział danych

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

### <a name='a5'></a>Budowa, uczenie, ocena perceptronu

```python
?MLPClassifier
```

```python
model = MLPClassifier((500, 200),
                      activation='relu',
                      solver='adam',
                      batch_size=200,
                      max_iter=1000,
                      shuffle=True,
                      early_stopping=True,
                      validation_fraction=0.05, #wielkość zbioru walidacyjnego
                      verbose=True)
```

```python
model.fit(X_train, y_train)
```

```python
accuracy_score(y_test, model.predict(X_test))
```
